```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.colors as colors
from matplotlib.colors import Normalize
```

# Mandelbrot set

```python
x = np.linspace(-1.5, 1.5, 1000)
y = np.linspace(-1.5, 1.5, 1000)
X,Y = np.meshgrid(x,y)

C = X+1j*Y
Z = C

for i in range(50):
  Z = Z*Z +C 


fig, ax = plt.subplots(figsize=(10,10))
ax.set_xlabel("Re(z)")
ax.set_ylabel("Im(z)")
ax.set_xticks([])
ax.set_yticks([])
plt.imshow(np.abs(Z)<2, cmap = 'gnuplot')
plt
```

#Tetration fractals

```python
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)
X,Y = np.meshgrid(x,y)

C = X+1j*Y
Z = C

for i in range(70):
  Z = Z**Z

plt.figure(figsize=(10,10))
ax.set_xlabel("Re(z)")
ax.set_ylabel("Im(z)")
ax.set_xticks([])
ax.set_yticks([])
plt.imshow(np.real(Z)<10**48, cmap = 'plasma')
plt.show()
```

#Newton's fractal

z³-1

```python
x = np.linspace(-1, 1, 1000)
y = np.linspace(-1, 1, 1000)
X,Y = np.meshgrid(x,y)

C = X+1j*Y
Z = C



for i in range(50):
  Z = Z-(Z**3-1)/(3*Z**2) 

filter1 = np.abs(Z-1j)<0.01
filter2 = np.abs(Z-((np.sqrt(3)/2)*1j-0.5))<0.01
filter3 = np.abs(Z-((-np.sqrt(3)/2)*1j-0.5))<0.01
z = np.zeros((1000,1000), dtype = 'int')
z[filter1] = 1
z[filter2] = 2
z[filter3] = 3

colors_list = ["blue","cyan","purple", "cyan"]
positions = [1, 2, 3]


cmap = colors.ListedColormap(colors_list)

fig, ax = plt.subplots(figsize=(10,10))
ax.set_xlabel("Re(z)")
ax.set_ylabel("Im(z)")
ax.set_xticks([])
ax.set_yticks([])
plt.imshow(z, cmap = cmap)

plt.show()
```

z⁴-1

```python
x = np.linspace(-1, 1, 1000)
y = np.linspace(-1, 1, 1000)
X,Y = np.meshgrid(x,y)

C = X+1j*Y
Z = C



for i in range(80):
  Z = Z-(Z**4-1)/(4*Z**3) 

filter1 = np.abs(Z-1j)<0.01
filter2 = np.abs(Z-1)<0.01 
filter3 = np.abs(Z+1j)<0.01
filter4 = np.abs(Z+1)<0.01


z = np.zeros((1000,1000), dtype = 'int')
z[filter1] = 1
z[filter2] = 2
z[filter3] =3
z[filter4] =4


colors_list = ["white", "red","green",  "yellow", "blue"]
positions = [1,  2,3, 4]


cmap = colors.ListedColormap(colors_list)
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_xlabel("Re(z)")
ax.set_ylabel("Im(z)")
ax.set_xticks([])
ax.set_yticks([])
plt.imshow(z,cmap = cmap)
plt.show()
```

z⁵+z²-z+1

```python
x = np.linspace(-3, 3, 1000)
y = np.linspace(-3, 3, 1000)
X,Y = np.meshgrid(x,y)

C = X+1j*Y
Z = C



for i in range(80):
  Z = Z-(Z**5+Z**2-Z+1)/(5*Z**4+2*Z-1) 

zeros  = np.roots([1, 0, 0, 1, -1, 1])
filter1 = np.abs(Z - zeros[0])  < 0.01
filter2 = np.abs(Z - zeros[1])  < 0.01
filter3 = np.abs(Z - zeros[2])  < 0.01
filter4 = np.abs(Z - zeros[3])  < 0.01
filter5 = np.abs(Z - zeros[4])  < 0.01


z = np.zeros((1000,1000), dtype = 'int')
z[filter1] = 1
z[filter2] = 2
z[filter3] =3
z[filter4] =4
z[filter5] = 5


colors_list = ["cyan", "Navy","cyan",  "RoyalBlue", "LightBlue", "SkyBlue"]
positions = [1,  2,3, 4, 5]


cmap = colors.ListedColormap(colors_list)
fig, ax = plt.subplots(figsize=(15, 15))
ax.set_xlabel("Re(z)")
ax.set_ylabel("Im(z)")
ax.set_xticks([])
ax.set_yticks([])
plt.imshow(z,cmap = cmap)
plt.show()
```
